```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Train.csv")
df.head()
```

```python
k, d, j = 0, [], 0
for i in range(len(df['Profession'])):
    if df['Profession'][i] in ['Engineer', 'Lawyer', 'Doctor'] and df['Graduated'][i]!='Yes':
        df['Graduated'][i]='Yes'
        j+=1
    if df['Profession'][i] not in d:
        d.append(df['Profession'][i])
        k += 1
print(k, j)
print(d)        
```

```python
df = df.replace({"Gender":{"Male":1, "Female":2}, 
            "Ever_Married":{"Yes":1, "No":0}, 
            "Graduated":{"Yes":1, "No":0},
            "Profession":{"Healthcare":1,"Engineer":2, "Lawyer":3, "Entertainment":4, "Artist":5, "Executive":6, "Doctor":7, "Homemaker":8, "Marketing":9},
            "Spending_score":{"Low":0, "Average":5, "High":10}, 
            "Var_1":{"Cat_1":1, "Cat_2":2, "Cat_3":3, "Cat_4":4, "Cat_5":5, "Cat_6":6, "Cat_7":7},
            })
df.head()
```

```python
df.isnull().sum()
```

```python
df['Work_Experience'].fillna(0, inplace=True)
df.head()
```

```python
df = df.dropna(axis=0)
df.head()
df.isnull().sum()
```

```python
import seaborn as sns

def plot_correlation_map(df):
    corr = df.corr()
    s , ax = plt.subplots( figsize =(12,10))
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    s = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 })
plot_correlation_map(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

x = df[['Age', 'Ever_Married', 'Graduated', 'Profession', 'Spending_Score', 'Family_Size']]
y = df['Segmentation']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=0)
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
y_pred = logreg.predict(x_test)
print("Accuracy = {:.2f}".format(logreg.score(x_test, y_test)))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=0)

knn = KNeighborsClassifier(n_neighbors=20)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
print('Accuracy = ', accuracy_score(y_pred, y_test))
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=10)
Tree = tree.DecisionTreeClassifier()
Tree.fit(x_train, y_train)
y_pred = Tree.predict(x_test)
print('Accuracy = {}'.format(accuracy_score(y_test, y_pred)))
```
