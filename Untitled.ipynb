```python
import math
def distance(d1, d2, l, metric='euc', q=2):
    if metric=='euc':#Eulidean distance
        return math.sqrt(sum([pow(d1[i]-d2[i], 2) for i in range(l)]))
    elif metric=='man':#Manhattan distance
        return sum([abs(d1[i]-d2[i]) for i in range(l)])
    elif metric=="min":#Minkowski distance
        return pow(sum([pow(abs(d1[i]-d2[i]), q) for i in range(l)]), 1/q)
    
import operator

def getNeighbors(trainingSet, testInstance, k, metric):
    distances = []
    length = len(testInstance)
    for x in range(len(trainingSet)):
        dist = distance(testInstance, trainingSet[x], length, metric)
        distances.append((trainingSet[x], dist)) #hseb les distances lkol
    distances.sort(key=operator.itemgetter(1))   #ratebhom
    neighbors = []
    for x in range(k):
        neighbors.append(distances[x][0])
    return neighbors                            #raja3 a9reb k-voisins

def getResponse(neighbors):
    classVotes={}
    for x in range(len(neighbors)):
        response = neighbors[x][-1]
        if response in classVotes:
            classVotes[response] += 1
        else:
            classVotes[response] = 1
    sortedVotes = sorted(classVotes.items(), key=operator.itemgetter(1), reverse=True)
    return sortedVotes[0][0]           #akther wehed dhhor fel classification

def main(tab, new, k, metric="man"):
    trainingSet = tab
    neighbors = getNeighbors(trainingSet, new, k, metric)
    return getResponse(neighbors)
t = [[1,3,5,4,6,1,1],
    [2,4,6,10,3,2,2],
    [3,8,3,4,2,6,3],
    [4,2,1,4,3,6,3],
    [5,2,5,1,4,8,2]]
print("Exercice 3 : \n   Question 1 :\n")
print("      Distance euclidienne et K=1 : ", main(t, [3, 12, 4, 7, 8], 1, "euc"))
print("      Distance Manhattan et K=1 : ", main(t, [3, 12, 4, 7, 8], 1, "man"))
print("      Distance Minkowski et K=1 : ", main(t, [3, 12, 4, 7, 8], 1, "min"))

print("\n      Distance euclidienne et K=3 : ",main(t, [3, 12, 4, 7, 8], 3, "euc"))
print("      Distance Manhattan et K=3 : ",main(t, [3, 12, 4, 7, 8], 3, "man"))
print("      Distance Minkowski et K=3 : ",main(t, [3, 12, 4, 7, 8], 3, "min"))

points = [[2, 6],[3, 1],[4, 2],[6, 0],[7, 5],[7, 3], [10, 3]]
target = [5, 5]

print("\n   Question 2 :\n\n      Le point le plus proche de (5,5) : ", getNeighbors(points, target, 1, "euc"))
print("      Les 3 points les plus proches de (5,5) : ", getNeighbors(points, target, 3, "euc"))
```
